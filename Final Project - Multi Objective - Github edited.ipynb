```python
def products(product):
    
    fillament_p = 2              # [Euro/100 gram]
    aluminium_p = 3              # [Euro/Kg]
    spring_p = 0.1               # [Euro]
    bolt_p = 0.02                # [Euro]
    nuts_p = 0.025               # [Euro]
    washer_p = 0.035             # [Euro]
    belt_p = 7                   # [Euro/m]
    fittings_p = 1               # [Euro]
    pulley_p = 3.5               # [Euro]
    carton_s_p = 0.5             # [Euro]
    carton_l_p = 0.7             # [Euro]
    
    cnc_milling_el = 1.5         # [KWh]
    cnc_thread_el = 0.75         # [KWh]
    fdm_el = 0.36                # [KWh]
    furnace_el = 2.5             # [KWh]
    press_el = 0                 # [KWh]
    packaging_el = 0.075         # [KWh]
    pooling_el = 0.34            # [KWh]
    
    el_grid_price = 0.187        # [Euro/KWh]
    salary = 1400                # [Euro]
    
    if product == 'Product_1':
# devices operating time (hour), water waste (liter)
        cnc_milling_t, cnc_milling_w = 1.8, 2.5    #[hour],[liter]           #[6][13]
        cnc_thread_t, cnc_thread_w = 0, 0          #[hour],[liter]           #[7][14]
        fdm_t = 2.5                                #[hour]                   #[8]
        furnace_t = 3                              #[hour]                   #[9]
        packaging_t = 0.1                          #[hour]                   #[10]
        pooling_t = 2                              #[hour]                   #[11]
        assembling_t = 0.6                         #[hour]                   #[12]
# amount of raw material
        fillament_a = 230/100                      #[gram]                   #[15]
        aluminium_a = 2.2                          #[kg]                     #[16]
        spring_a = 2                               #[number]                 #[17]
        bolt_a = 3                                 #[number]                 #[18]
        nuts_a = 3                                 #[number]                 #[19]
        washer_a = 0                               #[number]                 #[20]
        belt_a = 1                                 #[meter]                  #[21]
        fittings_a = 2                             #[number]                 #[22]
        pulley_a = 2                               #[number]                 #[23]
        carton_s_a = 1                             #[number]                 #[24]
        carton_l_a = 0                             #[number]                 #[25]
        selling_price = 50                         #[Euro]                   #[0]
     
    elif product == 'Product_2':
# devices operating time (hour), water waste (liter)
        cnc_milling_t, cnc_milling_w = 0, 0
        cnc_thread_t, cnc_thread_w = 2, 2
        fdm_t = 4
        furnace_t = 2
        packaging_t = 0.1
        pooling_t = 2.25
        assembling_t = 0.5
# amount of raw material
        fillament_a = 410/100
        aluminium_a = 2.2
        spring_a = 0
        bolt_a = 2
        nuts_a = 2
        washer_a = 0
        belt_a = 1.2
        fittings_a = 1
        pulley_a = 2
        carton_s_a = 1
        carton_l_a = 0
        selling_price = 75
        
    elif product == 'Product_3':
# devices operating time (hour), water waste (liter)
        cnc_milling_t, cnc_milling_w = 0.5, 1.8
        cnc_thread_t, cnc_thread_w = 0.75, 2.2
        fdm_t = 3.4
        furnace_t = 2.25
        packaging_t = 0.12
        pooling_t = 0
        assembling_t = 1
# amount of raw material
        fillament_a = 360/100
        aluminium_a = 2.5
        spring_a = 1
        bolt_a = 8
        nuts_a = 8
        washer_a = 0
        belt_a = 0
        fittings_a = 4
        pulley_a = 0
        carton_s_a = 0
        carton_l_a = 1
        selling_price = 50
        
    elif product == 'Product_4':
# devices operating time (hour), water waste (liter)
        cnc_milling_t, cnc_milling_w = 1.25, 2
        cnc_thread_t, cnc_thread_w = 0.5, 2
        fdm_t = 3.8
        furnace_t = 3.5
        packaging_t = 0.12
        pooling_t = 0
        assembling_t = 1
# amount of raw material
        fillament_a = 400/100
        aluminium_a = 2.2
        spring_a = 2
        bolt_a = 8
        nuts_a = 8
        washer_a = 0
        belt_a = 0
        fittings_a = 6
        pulley_a = 0
        carton_s_a = 0
        carton_l_a = 1
        selling_price = 50
    
    # material cost = amount of material * price of material
    material_cost = ((aluminium_a * aluminium_p) + (fillament_a * fillament_p) + (spring_a * spring_p)
                              + (bolt_a * bolt_p) + (nuts_a * nuts_p) + (washer_a * washer_p) + (belt_a * belt_p)
                              + (fittings_a * fittings_p) + (pulley_a * pulley_p) + (carton_s_a * carton_s_p)
                              + (carton_l_a * carton_l_p))
    
    # electricity cost = electricity grid price * device consumption * devices working time
    electricity_cost = el_grid_price * ((cnc_milling_t * cnc_milling_el) + (cnc_thread_t * cnc_thread_el)
                                + (fdm_t * fdm_el) + (furnace_t * furnace_el) + (pooling_t * pooling_el)
                                + (packaging_t * packaging_el))
    #[1]
    cost = material_cost + electricity_cost
    #[2]   
    time = cnc_milling_t + cnc_thread_t + fdm_t + furnace_t + packaging_t + pooling_t + assembling_t
    #[5]   
    water_waste = (cnc_milling_w * cnc_milling_t) + (cnc_thread_w * cnc_thread_t)
    
    worker_time = assembling_t + packaging_t + ((cnc_milling_t + cnc_thread_t + fdm_t + furnace_t + pooling_t) / 5)
    #[4]
    worker_salary = worker_time * salary/200
    #[3]
    gross_profit = selling_price - cost - worker_salary
    
    return [selling_price, cost, time, gross_profit, worker_salary, water_waste, cnc_milling_t, cnc_thread_t,
           fdm_t, furnace_t, packaging_t, pooling_t, assembling_t, cnc_milling_w, cnc_thread_w, fillament_a,
           aluminium_a, spring_a, bolt_a, nuts_a, washer_a, belt_a, fittings_a, pulley_a, carton_s_a, carton_l_a]
```

```python
def lpStatus(status):
    if status == 1:
        out = "Optimal"
    elif status == 0:
        out = "Not Solved"
    elif status == -1:
        out = "Infeasible"
    elif status == -2:
        out = "Unbounded"
    elif status == -3:
        out = "Undefined"
    else:
        out = status
    return out
```

<div id='fun_optimization'>
<br>
<h2>Function Definition: <code>Optimization</code></h2>
<br>

<p><code>Optimization</code> function optimizes both single and multiobjective functions.</p>
<p>The input parameters including:</p>
    <ul>
        <li><b>method</b>: <span style='color:blue'>'PROFIT'</span> - <span style='color:blue'>'SUSTAINABILITY'</span> - <span style='color:blue'>'WEIGHTED_SUM'</span> - <span style='color:blue'>'GOAL_PROGRAMMING'</span> - <span style='color:blue'>'CONSTRAINT_METHOD'</span></li>
        <li><b>shift</b>: number of working shift per day(company prefers one shift per day)</li>
        <li><b>a</b>: weighted number for multiobjective optimizations</li>
        <li><b>target_profit</b>: utopia value for objective function 1 (implements in goal programming)</li>
        <li><b>target_sustainability</b>: utopia value for objective function 2 (implements in goal programming)</li>
        <li><b>water_waste_constraint</b>: maximum amount of waste water as a constraint(implements in constraint method)</li>
    <ul>
<br>        
        </div>

```python
def Optimization(method, shift, a, target_profit, target_sustainability, water_waste_constraint):
    
################################################# INITIALIZE VARIABLES #################################################
    a0 = a
    a1 = 1 - a0

    pro_lst = [0] * len(products_lst)
    dev_lst = [0] * len(devices_lst)
    waste_water_eval = 0
    total_investment_eval = 0
    profit_eval = 0
    
############################################ DECLARING DECISIONAL VARIABLES #############################################
    
    # Decisional Variables for products
    for i in range(len(products_lst)):
        pro_lst[i] = p.LpVariable(name=products_lst[i], lowBound=min_demands[i], upBound=max_demands[i],cat='Integer')

    # Decisional Variables for devices
    for i in range(len(devices_lst)):
        dev_lst[i] = p.LpVariable(name=devices_lst[i], lowBound=0, cat='Integer')
        
    # Decisional Variables for number of workers
    worker = p.LpVariable(name='Worker', lowBound=0, cat='Integer')
    
############################################### SINGLE OBJECTIVE - PROFIT ###############################################
    
    if method == 'PROFIT':

    # Objective Function
        model = p.LpProblem('PROFIT', p.LpMaximize)

    # Objective Function
        obj = p.LpAffineExpression()
        # number of products * (selling price - production cost)
        for i in range(len(products_lst)):
            obj += pro_lst[i] * (products(products_lst[i])[0] - products(products_lst[i])[1])
        # number of devices * monthly service cost * shift number
        for i in range(len(devices_lst)):
            obj -= dev_lst[i] * devices(devices_lst[i])[2] * shift
        obj -= salary * worker
        obj *= (1 - tax)
        model += obj

############################################# SINGLE OBJECTIVE - SUSTAINABILITY #########################################
        
    elif method == 'SUSTAINABILITY':
        
    # Objective Function
        model = p.LpProblem('SUSTAINABILITY', p.LpMinimize)
        
    # Objective Function
        obj = p.LpAffineExpression()
        for i in range(len(products_lst)):
            obj += pro_lst[i] * products(products_lst[i])[5]
        model += obj
    

############################################# MULTI OBJECTIVE - WEIGHTED SUM #############################################
    
    elif method == 'WEIGHTED_SUM':
        
        model = p.LpProblem('WEIGHTED_SUM', p.LpMaximize)
        
    # Objective Function Profit
        obj = p.LpAffineExpression()
        # number of products * (selling price - production cost)
        for i in range(len(products_lst)):
            obj += a0 * pro_lst[i] * (products(products_lst[i])[0] - products(products_lst[i])[1])

        # number of devices * monthly service cost
        for i in range(len(devices_lst)):
            obj -= a0 * dev_lst[i] * devices(devices_lst[i])[2] * shift
        obj -= a0 * salary * worker
        obj *= (1 - tax)

    # Objective Function Sustainablity
        for i in range(len(products_lst)):
            obj -= a1 * pro_lst[i] * products(products_lst[i])[5] 
        model += obj   

########################################## MULTI OBJECTIVE - GOAL PROGRAMMING ############################################
    elif method == 'GOAL_PROGRAMMING':
        
        model = p.LpProblem('GOAL_PROGRAMMING', p.LpMinimize)
        
    # Objective Function Profit
        obj = p.LpAffineExpression()
        # number of products * (selling price - production cost)
        for i in range(len(products_lst)):
            obj += pro_lst[i] * (products(products_lst[i])[0] - products(products_lst[i])[1])

        # number of devices * monthly service cost
        for i in range(len(devices_lst)):
            obj -= dev_lst[i] * devices(devices_lst[i])[2] * shift
        obj -= salary * worker
        obj *= (1 - tax) * (-1 * a1)
        obj -= target_profit * (-1 * a1)

    # Objective Function Sustainablity
        for i in range(len(products_lst)):
            obj += a0 * pro_lst[i] * products(products_lst[i])[5]
        obj -= target_sustainability * a0
        model += obj
        
######################################### MULTI OBJECTIVE - CONSTRAINT METHOD ###########################################
    elif method == 'CONSTRAINT_METHOD':
        # Objective Function
        model = p.LpProblem('CONSTRAINT_METHOD', p.LpMaximize)

    # Objective Function
        obj = p.LpAffineExpression()
        # number of products * (selling price - production cost)
        for i in range(len(products_lst)):
            obj += pro_lst[i] * (products(products_lst[i])[0] - products(products_lst[i])[1])
        # number of devices * monthly service cost * shift number
        for i in range(len(devices_lst)):
            obj -= dev_lst[i] * devices(devices_lst[i])[2] * shift
        obj -= salary * worker
        obj *= (1 - tax)
        model += obj
        
        c9 = p.LpAffineExpression()
        for i in range(len(products_lst)):
            c9 += products(products_lst[i])[5] * pro_lst[i]
        model += c9 <= water_waste_constraint
        
######################################################### CONSTRAINTS ###################################################

    # Time Constraint for device 1 [CNC_milling]
    c1 = p.LpAffineExpression()
    for i in range(len(products_lst)):
        c1 += products(products_lst[i])[6] * pro_lst[i]
    model += c1 <= 200 * dev_lst[0] * shift

    # Time Constraint for device 2 [CNC_threading]
    c2 = p.LpAffineExpression()
    for i in range(len(products_lst)):
        c2 += products(products_lst[i])[7] * pro_lst[i]
    model += c2 <= 200 * dev_lst[1] * shift
    
    # Time Constraint for device 3 [3D_printer]
    c3 = p.LpAffineExpression()
    for i in range(len(products_lst)):
        c3 += products(products_lst[i])[8] * pro_lst[i]
    model += c3 <= 200 * dev_lst[2] * shift

    # Time Constraint for device 4 [Furnace]
    c4 = p.LpAffineExpression()
    for i in range(len(products_lst)):
        c4 += products(products_lst[i])[9] * pro_lst[i]
    model += c4 <= 200 * dev_lst[3] * shift
    
    # Time Constraint for device 5 [Packaging]
    c5 = p.LpAffineExpression()
    for i in range(len(products_lst)):
        c5 += products(products_lst[i])[10] * pro_lst[i]
    model += c5 <= 200 * dev_lst[4] * shift
    
    # Time Constraint for device 6 [Pooling]
    c6 = p.LpAffineExpression()
    for i in range(len(products_lst)):
        c6 += products(products_lst[i])[11] * pro_lst[i]
    model += c6 <= 200 * dev_lst[5] * shift

    # Investment Constraints for devices
    c7 = p.LpAffineExpression()
    for i in range(len(devices_lst)):
        c7 += devices(devices_lst[i])[1] * dev_lst[i]
    model += c7 <= max_investment

    # Number of workers constraint
    c8 = p.LpAffineExpression()
    for i in range(len(products_lst)):
        # Assembling Time + Packaging Time
        c8 += (products(products_lst[i])[12] + products(products_lst[i])[10]) * pro_lst[i]
        # 'CNC_milling' time = 0.2 worker time
        c8 += (products(products_lst[i])[6] * pro_lst[i]) / 5
        # 'CNC_threading' time = 0.2 worker time
        c8 += (products(products_lst[i])[7] * pro_lst[i]) / 5
        # '3D_printer' time = 0.2 worker time
        c8 += (products(products_lst[i])[8] * pro_lst[i]) / 5
        # 'Furnace' time = 0.2 worker time
        c8 += (products(products_lst[i])[9] * pro_lst[i]) / 5
        # 'Pooling' time = 0.2 worker time
        c8 += (products(products_lst[i])[11] * pro_lst[i]) / 5
    # devices working hours per month <= workers hours per month
    model += c8 <= worker * 200
    
######################################################### SOLVER ######################################################
    
    status = model.solve()
    
#################################################### SHOWING RESULTS ##################################################
    
    status_res = p.LpStatus[status]             #[0]
    obj_res = p.value(model.objective)          #[1]
    pro_1_val = p.value(pro_lst[0])             #[2]
    pro_2_val = p.value(pro_lst[1])             #[3]
    pro_3_val = p.value(pro_lst[2])             #[4]
    pro_4_val = p.value(pro_lst[3])             #[5]
    cnc_milling_val = p.value(dev_lst[0])       #[6]
    cnc_threading_val = p.value(dev_lst[1])     #[7]
    fdm_val = p.value(dev_lst[2])               #[8]
    furnace_val = p.value(dev_lst[3])           #[9]
    packaging_val = p.value(dev_lst[4])         #[10]
    pooling_val = p.value(dev_lst[5])           #[11]
    worker_val = p.value(worker)                #[12]
    
    # returns an lp file of created model
    model.writeLP(str(method)+".lp")

###################################################### EVALUATION ####################################################    

    for i in range(len(pro_lst)):
        # Showing Amount of waste water
        waste_water_eval += p.value(pro_lst[i]) * products(products_lst[i])[5]                                   #[13]
        profit_eval += p.value(pro_lst[i]) * (products(products_lst[i])[0] - products(products_lst[i])[1])

    for i in range(len(dev_lst)):
        # Showing Total Investment
        total_investment_eval += devices(devices_lst[i])[1] * p.value(dev_lst[i])                                #[14]
        profit_eval -= p.value(dev_lst[i]) * devices(devices_lst[i])[2] * shift
        
    profit_eval -= salary * p.value(worker)
    profit_eval *= (1 - tax)                                                                                     #[15]

    return [status_res, obj_res, pro_1_val, pro_2_val, pro_3_val, pro_4_val, cnc_milling_val, cnc_threading_val,
            fdm_val, furnace_val, packaging_val, pooling_val, worker_val, waste_water_eval, total_investment_eval,
           profit_eval]
```

<div id='fun_paretoFront'>
<br>
<h2>Function Definition: <code>paretoFront</code></h2>
<br>
<p><code>paretoFront</code> function implements a number for both Weighted-Sum and Goal Programming methods in a range between 0 and 1, by 1000 steps. Then calculates amount of profit, investment and waste water for each step. And it finds the unique values for profit and retrives all related parameters.</p><br>
    </div>

```python
def paretoFront(method, shift, target_profit, target_sustainability, water_waste_constraint):
    status_ls = list()            #[0]
    obj_ls = list()               #[1]
    pro_1_ls = list()             #[2]
    pro_2_ls = list()             #[3]
    pro_3_ls = list()             #[4]
    pro_4_ls = list()             #[5]
    cnc_milling_ls = list()       #[6]
    cnc_threading_ls = list()     #[7]
    fdm_ls = list()               #[8]
    furnace_ls = list()           #[9]
    packaging_ls = list()         #[10]
    pooling_ls = list()           #[11]
    worker_ls = list()            #[12]
    waste_water_ls = list()       #[13]
    total_investment_ls = list()  #[14]
    profit_ls = list()            #[15]
    tot_list = list()

    for i in np.arange(0,1.0001,0.001):
        results = Optimization(method, shift, i, target_profit, target_sustainability, water_waste_constraint)

        status_ls.append(results[0])
        obj_ls.append(results[1])
        pro_1_ls.append(results[2])
        pro_2_ls.append(results[3])
        pro_3_ls.append(results[4])
        pro_4_ls.append(results[5])
        cnc_milling_ls.append(results[6])
        cnc_threading_ls.append(results[7])
        fdm_ls.append(results[8])
        furnace_ls.append(results[9])
        packaging_ls.append(results[10])
        pooling_ls.append(results[11])
        worker_ls.append(results[12])
        waste_water_ls.append(results[13])
        total_investment_ls.append(results[14])
        profit_ls.append(results[15])
    
    for i in np.unique(profit_ls):
        indx = profit_ls.index(i)
        data_list = []
        data_list.append(status_ls[indx])
        data_list.append(obj_ls[indx])
        data_list.append(pro_1_ls[indx])
        data_list.append(pro_2_ls[indx])
        data_list.append(pro_3_ls[indx])
        data_list.append(pro_4_ls[indx])
        data_list.append(cnc_milling_ls[indx])
        data_list.append(cnc_threading_ls[indx])
        data_list.append(fdm_ls[indx])
        data_list.append(furnace_ls[indx])
        data_list.append(packaging_ls[indx])
        data_list.append(pooling_ls[indx])
        data_list.append(worker_ls[indx])
        data_list.append(waste_water_ls[indx])
        data_list.append(total_investment_ls[indx])
        data_list.append(profit_ls[indx])
        tot_list.append(data_list)
        
    return(method, profit_ls, total_investment_ls, waste_water_ls, tot_list)
```

```python
def paretoFrontPlot(method, profit_ls, total_investment_ls, waste_water_ls):
    plt.figure(figsize=(16,7))

    plt.subplot(1, 2, 1)
    plt.plot(profit_ls, total_investment_ls, color='black')
    plt.plot(profit_ls, total_investment_ls, 'o', color='red')
    plt.title(str(method), fontsize=16)
    plt.xlabel("Profit [Euro]", fontsize=12)
    plt.ylabel("Investment [Euro]", fontsize=12)

    plt.subplot(1, 2, 2)
    plt.plot(profit_ls, waste_water_ls, color='black')
    plt.plot(profit_ls, waste_water_ls, 'o', color='red')
    plt.title(str(method), fontsize=16)
    plt.xlabel("Profit [Euro]", fontsize=12)
    plt.ylabel("Waste Water [Litre]", fontsize=12)
    
    plt.show()
```

```python
def evaluation(N, tot_list):
    pro_1_lst = list()
    pro_2_lst = list()
    pro_3_lst = list()
    pro_4_lst = list()
    investment = 0
    service_cost = 0
    net_profit_lst = list()
    investment_lst = list()
    water_waste_lst = list()

    while len(pro_1_lst) <= N:
        pro_1 = random.randint(0, max_demands[0])
        pro_2 = random.randint(0, max_demands[1])
        pro_3 = random.randint(0, max_demands[2])
        pro_4 = random.randint(0, max_demands[3])

        cnc_milling_t = (pro_1 * products('Product_1')[6]) + (pro_2 * products('Product_2')[6]) +\
                        (pro_3 * products('Product_3')[6]) + (pro_4 * products('Product_4')[6])

        cnc_thread_t = (pro_1 * products('Product_1')[7]) + (pro_2 * products('Product_2')[7]) +\
                        (pro_3 * products('Product_3')[7]) + (pro_4 * products('Product_4')[7])

        fdm_t = (pro_1 * products('Product_1')[8]) + (pro_2 * products('Product_2')[8]) +\
                (pro_3 * products('Product_3')[8]) + (pro_4 * products('Product_4')[8])

        furnace_t = (pro_1 * products('Product_1')[9]) + (pro_2 * products('Product_2')[9]) +\
                    (pro_3 * products('Product_3')[9]) + (pro_4 * products('Product_4')[9])

        packaging_t = (pro_1 * products('Product_1')[10]) + (pro_2 * products('Product_2')[10]) +\
                    (pro_3 * products('Product_3')[10]) + (pro_4 * products('Product_4')[10])

        pooling_t = (pro_1 * products('Product_1')[11]) + (pro_2 * products('Product_2')[11]) +\
                    (pro_3 * products('Product_3')[11]) + (pro_4 * products('Product_4')[11])

        if cnc_milling_t == 0:
            cnc_milling_n = 0
        else:
            cnc_milling_n = (cnc_milling_t // 200) + 1

        if cnc_thread_t == 0:
            cnc_thread_n = 0
        else:
            cnc_thread_n = (cnc_thread_t // 200) + 1

        if fdm_t == 0:
            fdm_n = 0
        else:
            fdm_n = (fdm_t // 200) + 1

        if furnace_t == 0:
            furnace_n = 0
        else:
            furnace_n = (furnace_t // 200) + 1

        if packaging_t == 0:
            packaging_n = 0
        else:
            packaging_n = (packaging_t // 200) + 1

        if pooling_t == 0:
            pooling_n = 0
        else:
            pooling_n = (pooling_t // 200) + 1

        service_cost = (cnc_milling_n * devices('CNC_milling')[2]) + (cnc_thread_n * devices('CNC_threading')[2]) +\
                        (fdm_n * devices('3D_printer')[2]) + (furnace_n * devices('Furnace')[2]) +\
                        (packaging_n * devices('Packaging')[2]) + (pooling_n * devices('Pooling')[2])

        gross_profit = (pro_1 * products('Product_1')[3]) + (pro_2 * products('Product_2')[3]) +\
                        (pro_3 * products('Product_3')[3]) + (pro_4 * products('Product_4')[3])

        net_profit = (1 - tax) * (gross_profit - service_cost)

        investment = (cnc_milling_n * devices('CNC_milling')[1]) + (cnc_thread_n * devices('CNC_threading')[1]) +\
                    (fdm_n * devices('3D_printer')[1]) + (furnace_n * devices('Furnace')[1]) +\
                    (packaging_n * devices('Packaging')[1]) + (pooling_n * devices('Pooling')[1])

        water_waste = (pro_1 * products('Product_1')[5]) + (pro_2 * products('Product_2')[5]) +\
                        (pro_3 * products('Product_3')[5]) + (pro_4 * products('Product_4')[5])
        
        if investment <= max_investment:
            
            pro_1_lst.append(pro_1)
            pro_2_lst.append(pro_2)
            pro_3_lst.append(pro_3)
            pro_4_lst.append(pro_4)
            investment_lst.append(investment)
            net_profit_lst.append(net_profit)
            water_waste_lst.append(water_waste)

    plt.figure(figsize=(16,7))

    plt.subplot(1, 2, 1)
    plt.scatter(net_profit_lst, investment_lst, label='Random Numbers on feasible Region', color='orange')
    
    plt.plot([i[-1] for i in tot_list[-1]], [i[-2] for i in tot_list[-1]], label='Pareto-Front Points', marker = 'o',
           color='red')
    plt.plot([i[-1] for i in tot_list[-1]], [i[-2] for i in tot_list[-1]], label='Pareto-Front line', color='green')
    
    plt.title("Feasible Region", fontsize=16)
    plt.xlabel("Profit [Euro]", fontsize=12)
    plt.ylabel("Investment [Euro]", fontsize=12)
    plt.legend(loc="best", fontsize=12)

    plt.subplot(1, 2, 2)
    plt.scatter(net_profit_lst, water_waste_lst, label='Random Numbers on feasible Region', color='orange')
    
    plt.plot([i[-1] for i in tot_list[-1]], [i[-3] for i in tot_list[-1]], label='Pareto-Front Points', marker = 'o',
            color='red')
    plt.plot([i[-1] for i in tot_list[-1]], [i[-3] for i in tot_list[-1]], label='Pareto-Front line', color='green')
    plt.legend(loc="best", fontsize=12)
    
    plt.title("Feasible Region", fontsize=16)
    plt.xlabel("Profit [Euro]", fontsize=12)
    plt.ylabel("Waste Water [Litre]", fontsize=12)
    plt.show()
```
